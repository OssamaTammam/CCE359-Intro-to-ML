```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.under_sampling import RandomUnderSampler as underSampler
from sklearn.preprocessing import StandardScaler as scaler

```

```python
y = data[10]
x = data.drop(columns=10)
```
